```python
import hashlib
from PyPDF2 import PdfReader
```

```python
def hash_file(filename):
    """Return SHA-1 hash of the file."""
    h = hashlib.sha1()
    with open(filename, 'rb') as file:
        for chunk in iter(lambda: file.read(4096), b""):
            h.update(chunk)
    return h.hexdigest()

def extract_text(filename):
    """Extract text content from a PDF file."""
    reader = PdfReader(filename)
    text = ""
    for page in reader.pages:
        text += page.extract_text() or ""
    return text.strip()
```

```python
def compare_pdfs(file1, file2):
    # Step 1: Compare binary hashes
    hash1, hash2 = hash_file(file1), hash_file(file2)
    if hash1 == hash2:
        return "Files are completely identical (binary level)."
    
    # Step 2: Compare extracted text
    text1, text2 = extract_text(file1), extract_text(file2)
    if text1 == text2:
        return "Files differ in metadata/formatting, but text content is identical."
    else:
        return "Files are different in content."
```
