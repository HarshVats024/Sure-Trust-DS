```python
# Scientific Calculator - Harsh Vats under guidance of Purnangshu Roy Sir

import math

# Basic operations
def add(x, y): return x + y

def subtract(x, y): return x - y

def multiply(x, y): return x * y

def divide(x, y): return x / y if y != 0 else "Error: Division by zero."

def modulus(x, y): return x % y

# Advanced math
def power(x, y): return math.pow(x, y)

def square_root(x): return math.sqrt(x) if x >= 0 else "Error: Negative root."

def factorial(x): return math.factorial(int(x)) if x >= 0 and x == int(x) else "Error: Non-negative integers only."

def exponential(x): return math.exp(x)

def natural_log(x): return math.log(x) if x > 0 else "Error: Log undefined."

def log_base_10(x): return math.log10(x) if x > 0 else "Error: Log undefined."

# Trigonometry
def trig(func, x, deg=True):
    if deg: x = math.radians(x)
    return getattr(math, func)(x)

# Conversions
def to_base(x, base): return {2: bin, 8: oct, 16: hex}.get(base)(int(x))

def from_base(x, base): return int(x, base)

def angle_conversion(x, to): return math.radians(x) if to == 'r' else math.degrees(x)

def display_constants(): return {"pi": math.pi, "e": math.e}

def main_menu():
    print("\n=== Scientific Calculator ===")
    print("1. Basic Arithmetic\n2. Advanced Math\n3. Trigonometry\n4. Number Conversion\n5. Angle Conversion\n0. Exit")

def run_calculator():
    while True:
        main_menu()
        choice = input("\nSelect option: ").strip()

        if choice == '0':
            print("Goodbye!"); break

        elif choice == '1':
            x, y = float(input("x: ")), float(input("y: "))
            op = input("+ - * / %: ").strip()
            result = {'+': add, '-': subtract, '*': multiply, '/': divide, '%': modulus}.get(op, lambda *_: "Invalid")(x, y)
            print("Result:", result)

        elif choice == '2':
            print("1.Power 2.Root 3.Fact 4.Exp 5.Ln 6.Log10")
            ch = input("Select: ").strip()
            x = float(input("x: "))
            result = {
                '1': power(x, float(input("y: "))),
                '2': square_root(x),
                '3': factorial(x),
                '4': exponential(x),
                '5': natural_log(x),
                '6': log_base_10(x)
            }.get(ch, "Invalid")
            print("Result:", result)

        elif choice == '3':
            print("1.sin 2.cos 3.tan")
            f = {'1': 'sin', '2': 'cos', '3': 'tan'}.get(input("Select: ").strip())
            x = float(input("Angle: "))
            deg = input("Degrees? (y/n): ").lower() == 'y'
            print("Result:", trig(f, x, deg) if f else "Invalid")

        elif choice == '4':
            print("1.To Bin 2.Oct 3.Hex 4.From Bin 5.Oct 6.Hex")
            opt = input("Select: ").strip()
            val = input("Value: ").strip()
            try:
                result = {
                    '1': to_base(val, 2), '2': to_base(val, 8), '3': to_base(val, 16),
                    '4': from_base(val, 2), '5': from_base(val, 8), '6': from_base(val, 16)
                }.get(opt, "Invalid")
                print("Result:", result)
            except: print("Error: Invalid input")

        elif choice == '5':
            x = float(input("Angle: "))
            to = input("To radians (r) or degrees (d): ").lower()
            print("Result:", angle_conversion(x, to) if to in ['r', 'd'] else "Invalid")

        else:
            print("Invalid choice.")

if __name__ == "__main__":
    run_calculator()
```
